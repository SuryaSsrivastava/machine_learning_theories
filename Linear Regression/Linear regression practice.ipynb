```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X=np.arange(0,10)
Y=(X-5)**2
print(X,Y)
```

```python
plt.style.use("seaborn")
plt.plot(X,Y)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```

```python
x=10
lr=0.1
error=[]
plt.plot(X,Y)
for i in range(50):
    grad=2*(x-5)
    x=x-lr*grad
    e=(x-5)**2
    error.append(e)
    print(x)
    plt.scatter(x,e)
    
```

```python
plt.plot(error)
plt.show()
```
